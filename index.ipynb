## CLEANING


```python
travel_dataset['Travel Start Date']
```

```python
travel_dataset_copy = travel_dataset.copy() # copy dataset
```

```python
def remove_white_space(text):
    if type(text) == str:
        return text.strip()
    return text
```

```python
travel_dataset_copy['Travel Start Date'] = travel_dataset_copy['Travel Start Date'].apply(remove_white_space) # clearing whitespaces 
```

```python
travel_dataset.info()
```

```python
travel_dataset_copy.info()
```

```python
travel_dataset_copy['Travel Start Date'] = pd.to_datetime(travel_dataset_copy['Travel Start Date']) # changing travel start data from strings to date data type
```

```python
travel_dataset_copy.info()
```

```python
travel_dataset
```

```python
# change travel end date from string to datetime datatype
travel_dataset_copy['Travel End Date'] = travel_dataset_copy['Travel End Date'].apply(remove_white_space)
```

```python
travel_dataset_copy['Travel End Date'] = pd.to_datetime(travel_dataset_copy['Travel End Date'])
```

## null values in purpose of travel

```python
travel_dataset_copy[travel_dataset_copy['Purpose Of Travel'].isna()]
```

```python
travel_dataset_copy['Purpose Of Travel'] = travel_dataset_copy['Purpose Of Travel'].fillna('not provided')
```

```python
travel_dataset_copy.info()
```

## separate state and city

```python
def extract_state(text):
    results = text.split(',')[1]
    return results


def extract_city(text):
    results = text.split(',')[0]
    return results   
```

```python
travel_dataset_copy['state'] = travel_dataset_copy['Destination'].apply(extract_state)
travel_dataset_copy['city'] = travel_dataset_copy['Destination'].apply(extract_city)
```

```python
travel_dataset_copy
```

## separate name

```python
def extract_surname(text):
    results = text.split(' ')[0]
    return results

def extract_firstname(text):
    results = text.split(' ')[1]
    return results

```

```python
travel_dataset_copy['surname'] = travel_dataset_copy['Employee'].apply(extract_surname)
travel_dataset_copy['firstname'] = travel_dataset_copy['Employee'].apply(extract_firstname)
```
